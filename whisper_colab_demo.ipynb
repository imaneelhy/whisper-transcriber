```python
from google.colab import files

print("Choose an audio file (wav, mp3, m4a, etc.) from your computer...")
uploaded = files.upload()

# Get the first uploaded file name
audio_path = next(iter(uploaded.keys()))
print("Uploaded:", audio_path)
```

```python
from faster_whisper import WhisperModel

# You can try "tiny", "base", "small", "medium" later.
# "small" is a nice balance of speed/quality.
MODEL_SIZE = "small"

# If you enabled GPU runtime, you can try device="cuda"
# If not sure, leave device="cpu" – it will still work, just slower.
device = "cpu"  # change to "cuda" if you know you have GPU

print("Loading model, this may take a bit the first time...")
model = WhisperModel(MODEL_SIZE, device=device, compute_type="int8")

print(f"Transcribing: {audio_path}")
segments, info = model.transcribe(audio_path, beam_size=5)

print(f"\nDetected language: {info.language} (prob={info.language_probability:.2f})\n")
print("=== TRANSCRIPT ===\n")
full_text = []
for seg in segments:
    line = f"[{seg.start:6.2f}s -> {seg.end:6.2f}s] {seg.text}"
    print(line)
    full_text.append(seg.text)

print("\n=== FULL TEXT ONLY ===\n")
print(" ".join(full_text))
```
